# Reading the data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
df=pd.read_csv("C:\\Users\\USER\\Downloads\\Telco-Customer-Churn.csv")
df
```

# Checking for missing values

```python
df.isna().sum()
```

# Comparing churn count with respect to gender

```python
sns.countplot(x='gender',hue='Churn',data=df)
```

# Number of female senior citizens

```python
fm=df['gender']=='Female'
sc=df['SeniorCitizen']==1
No_count=df.loc[fm&sc,:]
No_count
sns.countplot(x='gender',hue='SeniorCitizen',data=No_count)
```
